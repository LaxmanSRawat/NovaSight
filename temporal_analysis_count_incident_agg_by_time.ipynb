### Import Dependencies

```python
import pandas as pd
from datetime import datetime
```

### Load Parquet Files

```python
crime_parquet_path = "src/nyc_data_crime.parquet"
potential_parquet_path = "src/nyc_data_potential_crime.parquet"
crime_df = pd.read_parquet(crime_parquet_path)
potential_crime_df = pd.read_parquet(potential_parquet_path)
```

```python
crime_df.head()
```

```python
potential_crime_df.head()
```

```python
crime_df_grouped_by_typ_incident_date = crime_df.groupby(['typ_desc','incident_date']).count()['cad_evnt_id'].rename('count', inplace=True)

potential_crime_df_grouped_by_typ_incident_date = potential_crime_df.groupby(['typ_desc','incident_date']).count()['cad_evnt_id'].rename('count', inplace=True)
```

```python
crime_df_grouped_by_typ_incident_time = crime_df.groupby(['typ_desc','incident_time']).count()['cad_evnt_id'].rename('count', inplace=True)
potential_crime_df_grouped_by_typ_incident_time = potential_crime_df.groupby(['typ_desc','incident_time']).count()['cad_evnt_id'].rename('count', inplace=True)
```

### Export file as csv

```python
crime_df_grouped_by_typ_incident_date.to_csv('src/nyc_count_crime_type_by_date.csv', index=True)
potential_crime_df_grouped_by_typ_incident_date.to_csv('src/nyc_count_potential_crime_type_by_date.csv', index=True)
```

```python
crime_df_grouped_by_typ_incident_time.to_csv('src/nyc_count_crime_type_by_time.csv', index=True)
potential_crime_df_grouped_by_typ_incident_time.to_csv('src/nyc_count_potential_crime_type_by_time.csv', index=True)
```
