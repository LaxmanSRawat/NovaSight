### Import Dependencies

```python
import pandas as pd
from datetime import datetime
```

### Load Parquet Files

```python
crime_parquet_path = "src/nyc_data_crime.parquet"
potential_parquet_path = "src/nyc_data_potential_crime.parquet"
crime_df = pd.read_parquet(crime_parquet_path)
potential_crime_df = pd.read_parquet(potential_parquet_path)
```

```python
crime_df.head()
```

```python
potential_crime_df.head()
```

### Calculating the duration of call

```python
crime_df['call_duration_minutes'] = (pd.to_datetime(crime_df['closng_ts']) - pd.to_datetime(crime_df['incident_date'] + ' ' + crime_df['incident_time'])).dt.total_seconds() / 60.0
```

```python
crime_df.loc[:,['incident_date','incident_time','closng_ts','call_duration_minutes']].head(10)
```

```python
potential_crime_df['call_duration_minutes'] = (pd.to_datetime(potential_crime_df['closng_ts']) - pd.to_datetime(potential_crime_df['incident_date']+ ' ' + potential_crime_df['incident_time'])).dt.total_seconds() / 60.0
```

```python
potential_crime_df.loc[:,['incident_date','incident_time','closng_ts','call_duration_minutes']].head(10)
```

### Calculating top 10 category by call duration

```python
crime_category_by_call_duration = crime_df.groupby('typ_desc')['call_duration_minutes'].median().sort_values(ascending=False)
```

```python
crime_category_by_call_duration.head(20)
```

```python
potential_crime_category_by_call_duration = potential_crime_df.groupby('typ_desc')['call_duration_minutes'].median().sort_values(ascending=False)
```

```python
potential_crime_category_by_call_duration.head(20)
```

### Slicing data for top 10 category for visualization in d3

```python
crime_categories = crime_category_by_call_duration.head(10).index.tolist()
```

```python
crime_categories_df = crime_df[crime_df['typ_desc'].isin(crime_categories)]
crime_categories_df.shape
```

```python
crime_categories_df.head()
```

```python
potential_crime_categories = potential_crime_category_by_call_duration.head(10).index.tolist()
```

```python
potential_crime_categories_df = potential_crime_df[potential_crime_df['typ_desc'].isin(potential_crime_categories)]
potential_crime_categories_df.shape
```

```python
crime_categories_df.head()
```

### Export file as csv

```python
crime_categories_df.to_csv('src/nyc_data_crime_top_10_categories_by_call_duration.csv', index=False)
potential_crime_categories_df.to_csv('src/nyc_data_potential_crime_top_10_categories_by_call_duration.csv', index=False)
```

### Saving the bottom 10 category by average total duration

```python
crime_category_by_call_duration_b10 = crime_df.groupby('typ_desc')['call_duration_minutes'].median().sort_values(ascending=True)
potential_crime_category_by_call_duration_b10 = potential_crime_df.groupby('typ_desc')['call_duration_minutes'].median().sort_values(ascending=True)

crime_categories_b10 = crime_category_by_call_duration_b10.head(10).index.tolist()
potential_crime_categories_b10 = potential_crime_category_by_call_duration_b10.head(10).index.tolist()

crime_categories_df_b10 = crime_df[crime_df['typ_desc'].isin(crime_categories_b10)]
potential_crime_categories_df_b10 = potential_crime_df[potential_crime_df['typ_desc'].isin(potential_crime_categories_b10)]
```

```python
crime_categories_df_b10.to_csv('src/nyc_data_crime_bottom_10_categories_by_call_duration.csv', index=False)
potential_crime_categories_df_b10.to_csv('src/nyc_data_potential_crime_bottom_10_categories_by_call_duration.csv', index=False)
```
